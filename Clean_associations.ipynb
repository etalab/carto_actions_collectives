# Clean asso

```python
associations_ref = pd.read_csv("associations_amelie.csv", sep=";")
```

```python
associations_ref.head()
```

```python
associations_ref
```

```python
associations_ref.columns = ['Commitments', 'Associations']
```

```python
# On enlève les numéros : 1), 2), ...
for i in range(0, 50):
    associations_ref.Associations = associations_ref.Associations.str.replace(str(i) + '\) ', '')
    #associations_ref.Commitments = associations_ref.Commitments.str.replace(str(i), '').str.replace('.', '')

# On drop un titre de colonne restant ("Climate and sustainable...")
associations_ref.drop(9, axis=0, inplace=True)
# On ajoute un '#' pour le style.
associations_ref.Commitments = ['#' + x for x in associations_ref.Commitments]
```

```python
associations_ref.head()
```

```python
associations_ref.to_csv('/Users/gowitheflow/Documents/ETALAB/carto_amelie/cartographie/associations_clean.csv', sep=';', encoding='utf-8', index=False)
```

```python
associations_ref.Associations = associations_ref.Associations.str.replace('\n', '; ')
```

```python
associations_ref.fillna('Aucune association', inplace=True)
associations_ref.head(2)
```

```python
associations_dict = associations_ref.set_index('Commitments')['Associations'].to_dict()
```

```python
associations_dict
```

```python
import json
with open('associations_dict.json', 'w') as fp:
    json.dump(associations_dict, fp)
```
